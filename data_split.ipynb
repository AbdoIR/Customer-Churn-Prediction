<h1>Data spliting</h1>

<p>Welcome mfers to a data splitting masterclass buckle up and witness greatness !!!!!</p>

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# dataset loaded bitches 
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# then it splits perfectly because I'm doing it 
# random_state ensures that any of you dumasses will get the same results when running this code 
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# saved to csv like a Bossssssss
train_df.to_csv('train_data.csv', index=False)
test_df.to_csv('test_data.csv', index=False)

print(f"Train set saved with {len(train_df)} rows.")
print(f"Test set saved with {len(test_df)} rows.")


#signed by aymene
```
