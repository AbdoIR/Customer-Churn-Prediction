<h1>Data spliting</h1>

<p>Welcome mfers to a data splitting masterclass buckle up and witness greatness !!!!!</p>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# data loaded mfers
df = pd.read_csv('clean-database.csv')

# Performed 80/20 Split like a boss making sure that even if you rerun the code you'll get same results 
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Saving to CSV as it should be 
train_df.to_csv('train_data.csv', index=False)
test_df.to_csv('test_data.csv', index=False)

print("Split complete.")
```
